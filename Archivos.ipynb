#**Funciones**





## Uso


En Python, una función agrupa declaraciones relacionadas para que realicen una tarea específica.​

Las funciones ayudan a dividir un programa en partes más pequeñas y modulares manteniéndolo  organizado y manejable.​

Además,  el uso de funciones evita la repetición y hace que el código sea reutilizable.

Funciones.png

## Llamado a funciones

La instrucción **return** en Python termina la ejecución de la función como tal y le devuelve el control a la función de llamada o etapa donde fue llamada la función para que continue con el flujo normal de código.


Para usar a una función, simplemente se escribe el nombre de la función con los parámetros apropiados entre paréntesis separados por comas.​

---

***nombre_func***(param1,param2)

Ejemplos

```python
def suma(a, b):
  """Esta función realiza la suma: c = a + b​
  sus llamado se hace​
  suma(a,b)"""  
  return a + b

c=suma(13,2)
print("la suma es",c)
```

```python
# función sin parámetros o retorno de valores
def Hola():
  print("Hello!")

Hola()  # llamada a la función, 'Hello!' se muestra en la consola
```

```python
# función con un parámetro
def holaConNombre(name):
  print("Hello " + name + "!")

holaConNombre("Ada")  # llamada a la función, 'Hello Ada!' se muestra en la consola

```

## Parámetros por defecto ​

Python permiten definir nuestras funciones con parámetros por defecto. 



```python
def resta(a = None, b = None):
  """Esta función realiza la resta: c = a + b​ sus llamado se hace  resta(a,b)"""
  if a == None or b == None:
    print("Error, debes enviar dos números a la función")
    return
  print("realizando la resta")
  return a - b
```

```python
resta()
```

Al llamar a una función Python permite pasar los parámetros en cualquier orden  utilizando el nombre del parámetro.



```python
resta( 3, 2)
```

```python
resta( b = 3, a = 2)
```

```python
result = resta(3, b=2)
#result = 5
```

```python
result2 = suma(a=2,3)
```

## Alcance de las variables

Las variables definidas dentro de una función solo existen dentro de esa función.

```python
def duplica(num):
    x = num * 2
    return x

print(x)  # error - x no está definida
```

```python
print(duplica(4))  # muestra 8
```
