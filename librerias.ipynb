## Manejo de excepciones

Para manejar una excepción en Python se  usa la sentencia **try**. Cuando python detecta la sentencia **try**  estará atento a uno o más tipos diferentes de excepciones (según cómo lo indique), y si se generan, le permitirá reaccionar.

**try:**

>   Intenta ejecutar este código.

**except** <tipo_de_excepción>:

>  Si ocurre una excepción de este tipo, detén el proceso inmediatamente y salta a este bloque de código.

**finally:** 

>	 Esto se realiza si existe o no una excepción 


**Ejemplo:**

```python
while True:
    try:
        x = int(input("Ingrese un número: "))
        break
    except ValueError:
        print("Oops!  No ingreso un numero válido.  Trate de nuevo...")
```



*   Se recomienda mantener el bloque de prueba lo mas corto posible conteniendo solo el código que puede causar las excepciones que desea manejar. 
*   Las cláusulas de excepción deben ser lo más específicas porsibles.

*   Se debe asegurar que el código maneje correctamente tanto los errores esperados como los inesperados.

# Archivos

### Abrir archivos 

En Python la función **open()** permite abrir un determinado archivo, La función recibe dos argumentos: el primer argumento es el nombre del archivo, el segundo es el modo de apertura que indica cómo se abrirá el archivo:

*   **r**: Abre el archivo en modo lectura.
*   **r+**: Abre el archivo  en modo leectura y escritura.
*   **w**: Sobreescribir el contenido del archivo.
*   **a**: Si el achivo ya existe, añade el contenido al final.

La función open devuelve lo que se conoce como file handle, y es dado por el sistema operativo. Una vez que se ha terminado de usar este file handle es importante devolverlo y cerrarlo. 

### Leer archivos 

A continuación se presentan tres funciones para leer un archivo en Python:

#### **read()**

 La función *read()* retorna todo el contenido del archivo como una cadena de caracteres.

```python
try:
  f = open("nombres.txt")
  contenido=f.read()
  print(type(contenido))
  print(contenido)
finally:
        f.close()
```

Los gestores de contexto son estructuras en Python que te ayudarán muchísimo al trabajar con archivos.

**with open**( "\<archivo\>", "\<modo\>" ) **as** *\<var\>*

```python
with open("nombres.txt") as f:
    print(f.read(5))
aux="final"
```

1. Programa para calcular las raíces (r1 y r2) de la ecuación de segundo grado $ax²+bx+c = 0 $ dados los tres coeficientes a, b y c.​

```python
import math 
#Se piden los datos
a =float( input("Ingresa a: "))
b =float( input("Ingresa b: "))
c =float( input("Ingresa c: "))

#Caulculos
D=(b**2)-(4*a*c)
if D== 0:
  r1=-b / (2 * a)
  r2=r1
elif D>0:
    r1=(b+math.sqrt(D))/(2*a)
    r2=(b-math.sqrt(D))/(2*a)
else:
  pr=b/(2*a)
  pi=pow(D,.5)/(2*a)
  r1=complex(pr,pi)
  r2=complex(pr,-pi) 

#Se presentan resultados 

print("Las raices son:  ")
print("r1 = ",r1 )
print("r2 = ",r2 )
```

2. Programa que genera 10 números aleatorios  y los registra en un archivo con la fecha y hora de su generación llamado. El archivo generado se denomina *random.txt*

```python
from datetime import datetime
import time
import random


with open("random.txt","w+") as f:
  for i in range(0,10):
    value=random.randint(1,100)
    print(datetime.now().strftime('%Y/%m/%d_%H:%M:%S - '),value)
    f.write(datetime.now().strftime('%Y/%m/%d_%H:%M:%S - '))
    time.sleep(1)
    f.write(str( value ))
    f.write("\n")
```

3. Programa que pide al usuario el día y mes de su nacimiento para calcular el número de días faltan para su próximo cumpleaños.

```python
from datetime import date
today = date.today()


birth_day=int(input("¿En que día naciste (1 - 31)?"))
birth_month=int(input("¿En que mes naciste (1 - 12)?"))

my_birthday= date(today.year,birth_month,birth_day)
if my_birthday < today:
  my_birthday = my_birthday.replace(year=today.year+1)

print("El día de hoy es:", today)
print("La fecha de tu próximo cumpleaños es:", my_birthday)
time_to_birthday = abs(my_birthday-today)
print(" Falta ", time_to_birthday,"para tu cumpleaños")

```
