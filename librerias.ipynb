## Manejo de excepciones

Para manejar una excepcin en Python se  usa la sentencia **try**. Cuando python detecta la sentencia **try**  estar atento a uno o ms tipos diferentes de excepciones (segn cmo lo indique), y si se generan, le permitir reaccionar.

**try:**

>   Intenta ejecutar este cdigo.

**except** <tipo_de_excepcin>:

>  Si ocurre una excepcin de este tipo, detn el proceso inmediatamente y salta a este bloque de cdigo.

**finally:** 

>	 Esto se realiza si existe o no una excepcin 


**Ejemplo:**

```python
while True:
    try:
        x = int(input("Ingrese un nmero: "))
        break
    except ValueError:
        print("Oops!  No ingreso un numero vlido.  Trate de nuevo...")
```



*   Se recomienda mantener el bloque de prueba lo mas corto posible conteniendo solo el cdigo que puede causar las excepciones que desea manejar. 
*   Las clusulas de excepcin deben ser lo ms especficas porsibles.

*   Se debe asegurar que el cdigo maneje correctamente tanto los errores esperados como los inesperados.

### Abrir archivos 

En Python la funcin **open()** permite abrir un determinado archivo, La funcin recibe dos argumentos: el primer argumento es el nombre del archivo, el segundo es el modo de apertura que indica cmo se abrir el archivo:

*   **r**: Abre el archivo en modo lectura.
*   **r+**: Abre el archivo  en modo leectura y escritura.
*   **w**: Sobreescribir el contenido del archivo.
*   **a**: Si el achivo ya existe, aade el contenido al final.

La funcin open devuelve lo que se conoce como file handle, y es dado por el sistema operativo. Una vez que se ha terminado de usar este file handle es importante devolverlo y cerrarlo. 

### Leer archivos 

A continuacin se presentan tres funciones para leer un archivo en Python:

1. Programa para calcular las races (r1 y r2) de la ecuacin de segundo grado $ax+bx+c = 0 $ dados los tres coeficientes a, b y c.

```python
import math 
#Se piden los datos
a =float( input("Ingresa a: "))
b =float( input("Ingresa b: "))
c =float( input("Ingresa c: "))

#Caulculos
D=(b**2)-(4*a*c)
if D== 0:
  r1=-b / (2 * a)
  r2=r1
elif D>0:
    r1=(b+math.sqrt(D))/(2*a)
    r2=(b-math.sqrt(D))/(2*a)
else:
  pr=b/(2*a)
  pi=pow(D,.5)/(2*a)
  r1=complex(pr,pi)
  r2=complex(pr,-pi) 

#Se presentan resultados 

print("Las raices son:  ")
print("r1 = ",r1 )
print("r2 = ",r2 )
```

2. Programa que genera 10 nmeros aleatorios  y los registra en un archivo con la fecha y hora de su generacin llamado. El archivo generado se denomina *random.txt*

```python
from datetime import datetime
import time
import random


with open("random.txt","w+") as f:
  for i in range(0,10):
    value=random.randint(1,100)
    print(datetime.now().strftime('%Y/%m/%d_%H:%M:%S - '),value)
    f.write(datetime.now().strftime('%Y/%m/%d_%H:%M:%S - '))
    time.sleep(1)
    f.write(str( value ))
    f.write("\n")
```

3. Programa que pide al usuario el da y mes de su nacimiento para calcular el nmero de das faltan para su prximo cumpleaos.

```python
from datetime import date
today = date.today()


birth_day=int(input("En que da naciste (1 - 31)?"))
birth_month=int(input("En que mes naciste (1 - 12)?"))

my_birthday= date(today.year,birth_month,birth_day)
if my_birthday < today:
  my_birthday = my_birthday.replace(year=today.year+1)

print("El da de hoy es:", today)
print("La fecha de tu prximo cumpleaos es:", my_birthday)
time_to_birthday = abs(my_birthday-today)
print(" Falta ", time_to_birthday,"para tu cumpleaos")

```
