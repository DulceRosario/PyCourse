# Listas –append()

La función **append()** agrega un elemento al final de la lista

Se crea una variable **lista_listas** con dos elementos de tipo lista

```python
lista_listas=[ [1,2,3,4], ["a","b","s","e"] ]
print(lista_listas)
```

Con la función append() se agrega un nuevo elemento al segundo elemento de la variable **lista_listas** 

```python
lista_listas[1].append("Z")
print(lista_listas)
```

```python
lista_listas.append("Z")
print(lista_listas)
```

# range

  **range(n,m,p)** es un tipo de datos que contiene una sucesión aritmética de números enteros, donde:​
  * **n** indica el entero inicial de la sucesión​
  * **m** indica el entero final de la sucesión menos 1, es decir   m-1 será el último valor​
  * **p** define el incremento de los elementos de la sucesión ​

Si sólo se escriben dos argumentos dentro de **range(n,m)**, a p se le asigna el valor de 1, es decir la sucesión irá incrementándose de uno en uno​ **range(n, m)** es lo mismo que **range(n, m, 1)**.​

  Si se escribe sólo un argumento **range(m)**, entonces a n se le da el valor de 0 y a p el valor de 1. ​**range(m)** es lo mismo que **range(0, m, 1)**.

```python
range(1,9)
```

```python
a=range(4)
type(a)
```

Para ver los valores de un range es necesario convertirlos a lista con la función **list( )**.

```python
list(range(1,8))
```

```python
list(range(1,10,2))
```

```python
list(range(-3,10))
```

# For

**for** Permite ejecutar un conjunto de instrucciones un número determinado de veces. En Python el **for itera** sobre los elementos de cualquier **secuencia** (listas, cadenas, etc.) y en el orden en que aparecen

fordesc.png

## Ejemplos

```python
dias=["lunes", "martes", "miercoles", "jueves", "viernes", "sábado", "domingo"]
for elemento in dias:
  print(elemento)
```

```python
dias=["lunes", "martes", "miercoles", "jueves", "viernes", "sábado", "domingo"]
naturales = [0,1,2,3,4,5,6,7,8] 
lista_con_listas=[dias,naturales,["cadena",2]] 
long=len(lista_con_listas)
print( lista_con_listas[long-1] )
for i in lista_con_listas:
    print("elemento", i)
```

Observe que en el siguiente código, se generan las posiciones de los elementos usando la función range y el número de elemenos:

```python
surnames = ['Rivest', 'Shamir', 'Adleman'] 
for position in range(len(surnames)): 
    print(position, surnames[position]) 
```

**For** puede iterar sobre cualquier secuencia o colección, por lo que no es necesario obtener la lista de posiciones y recuperar elementos de una secuencia en cada iteración.

```python
surnames = ['Rivest', 'Shamir', 'Adleman'] 
long=len(surnames)
for position in range(long): 
    print(position, surnames[position]) 
```

```python
surnames = ['Rivest', 'Shamir', 'Adleman'] 
for surname in surnames: 
    print(surname) 
```

Programa que divide una lista de números en pares e impares

```python
numeros=[1,2,45,64,2,3,688,4,3,57,97]
pares=[]
impares=[]
for num in numeros:
  if num%2==0:
    pares.append(num)
  else:
    impares.append(num)

print("Los números pares son: ", pares)

print("Los números impares son: ", impares)


```

# BREAK y CONTINUE

Estas sentencias de utilizan para terminar la iteración actual e incluso todo el ciclo sin verificar la condición de paro.​

La sentencia **break** termina el bucle que la contiene. Si la instrucción break está dentro de un ciclo anidado se terminará el bucle más interno​

La sentencia **continue** se usa para omitir el resto del código dentro de un ciclo solo para la iteración actual. El ciclo no termina sino que continúa con la siguiente iteración.

## Ejemplos

Programa para encontrar los primeros 5 múltiplos de 6

```python
i = 1
while i <= 10:
    print('6 * ',(i), '=',6 * i)
    if i >= 5:
        break
    
    i = i + 1
```

Programa que imprime los números del 0 al 4 imitiendo el 3

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```
