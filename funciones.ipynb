# **Funciones con Listas**

**append**
Agregar un elemento a una lista (al final)

```python
mylist = ['uno', 'dos', 'tres', 'cuatro', 'cinco']

mylist.append(3)
mylist.append("3")


print(mylist)
```

**count**

Devuelve el número de veces que la lista contiene el objeto obj

```python
mylist = list("esteganografia")

print(mylist.count("a"))

```

**extend**

 Agrega los elementos de  seq a la  lista

```python
mylist = ['uno', 'dos', 'tres', 'cuatro', 'cinco']

list2 = ["Hola", "amigo"]

mylist.extend(list2)

print(mylist)
```

El código anterior es equivalente a usar el operador **+**

```python
mylist = ['uno', 'dos', 'tres', 'cuatro', 'cinco']

list2 = ["Hola", "amigo"]



mylist2= mylist +list2

print(mylist)
print(mylist2)

```

**index**

Regresa el  primer índice de la lista  en donde aparece el elemento obj por primera vez

```python
mylist = ['tres','uno', 'dos', 'cuatro', 'cinco', 'tres', 'tres']
ind=mylist.index('tres')
print(ind)
```

**sort**

Ordena los elementos de la lista​

```python
mylist = ['cde', 'fgh', 'abc', 'klm', 'opq'] 
list2 = [3, 5, 2, 4, 1]

mylist.sort() # En el caso de string ordena alfabeticamente


list2.sort()

print(mylist)

print(list2)
```

**insert**

Inserta el elemento obj dentro de la lista en el índice indicado

```python
mylist = [8, 2, 3, 4, 5]

mylist.insert(0,'Hello')

print(mylist)
```

**reverse**

Invierte los elementos de la lista 

```python
mylist = [1, 6, 9, 4, 5]

mylist.reverse()

print(mylist)
```

**pop**

Remueve  y regresa el  último elemento  obj de  la lista list

```python
mylist = ['one', 'two', 'three', 'four', 'five']

removed = mylist.pop()

print(mylist)

print(removed)
```

```python
mylist = ['one', 'two', 'three', 'four', 'five']

removed = mylist.pop(1)

print(mylist)

print(removed)
```

**Remove**

Elimina el elemento obj de list​

```python
mylist = ['one', 'two', 'three', 'four', 'five','two']

mylist.remove('two') #Elimina el primer valor encontrado

print(mylist)
```

## Funciones recursivas 

Una función recursiva es aquella se llama a sí misma para producir un resultado.

El ejemplo más común de una función recursiva es el factorial.

image.png

```python
def factorial(n):
  print("Calculando el factorial de ", n)
  if n in (0, 1):  # caso base
    return 1
  return factorial(n - 1) * n  # caso recursivo
```

Una función recursiva se puede escribir de manera iterativa siendo esta última la versión que recomiendan los programadores:

```python
def factorial(n):
    factorial = 1
    for i in range(n):
        factorial *= (i + 1)
    return factorial
```

### Datetime

Proporciona clases para manipular fechas y horas.

**Ejemplo**

```python
from datetime import date
today = date.today()


birth_day=int(input("¿En que día naciste (1 - 31)?"))
birth_month=int(input("¿En que mes naciste (1 - 12)?"))

my_birthday= date(today.year,birth_month,birth_day)
if my_birthday < today:
  my_birthday = my_birthday.replace(year=today.year+1)

print("El día de hoy es:", today)
print("La fecha de tu próximo cumpleaños es:", my_birthday)
time_to_birthday = abs(my_birthday-today)
print(" Falta ", time_to_birthday,"para tu cumpleaños")

```
