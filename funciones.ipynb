## Funciones recursivas 

Una funcin recursiva es aquella se llama a s misma para producir un resultado.

El ejemplo ms comn de una funcin recursiva es el factorial.

image.png

```python
def factorial(n):
  print("Calculando el factorial de ", n)
  if n in (0, 1):  # caso base
    return 1
  return factorial(n - 1) * n  # caso recursivo
```

Una funcin recursiva se puede escribir de manera iterativa siendo esta ltima la versin que recomiendan los programadores:

```python
def factorial(n):
    factorial = 1
    for i in range(n):
        factorial *= (i + 1)
    return factorial
```

### Datetime

Proporciona clases para manipular fechas y horas.

**Ejemplo**

```python
from datetime import date
today = date.today()


birth_day=int(input("En que da naciste (1 - 31)?"))
birth_month=int(input("En que mes naciste (1 - 12)?"))

my_birthday= date(today.year,birth_month,birth_day)
if my_birthday < today:
  my_birthday = my_birthday.replace(year=today.year+1)

print("El da de hoy es:", today)
print("La fecha de tu prximo cumpleaos es:", my_birthday)
time_to_birthday = abs(my_birthday-today)
print(" Falta ", time_to_birthday,"para tu cumpleaos")

```
